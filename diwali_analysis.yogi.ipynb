IMPORT NACCESARY LIBRARY

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("Diwali Sales Data.csv",encoding= 'unicode_escape')
df.shape
```

drop columns

```python
df=df.drop(columns=["Status","unnamed1"])
df
```

find null values

```python
df.isna().sum()
```

```python
df=df.dropna(how="any")
df
```

```python
df.info()
```

```python
df["Amount"]=df["Amount"].astype(int)
```

GENDER 

```python
ax=sns.countplot(data=df,x="Gender")
for num in ax.containers:
    ax.bar_label(num)
plt.show()
```

```python
df.columns
```

```python
ax=sns.countplot(data=df,x="Gender",hue="Marital_Status")
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

AGE_GROUP 

```python
df["Age Group"]
```

```python
ax=sns.countplot(data=df,x="Age Group")
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

UP state and group age

```python
sdf=df.loc[df["State"]=="Uttar Pradesh"]
sdf
sns.countplot(data=sdf,x="Age Group")
plt.title("age group sales in UP")
plt.show()
```

STATE WISE

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df,y="State")
plt.xticks(rotation=65)
plt.show()
```

Marital_Status

```python
ax=sns.countplot(data=df,x="Marital_Status",hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
df
```

HEALTHCARE

```python
ndf=df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(data=ndf,x="Occupation",y="Amount")
plt.show()
```

```python
odf=df.groupby(["Occupation"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(data=odf,x="Occupation",y="Orders")
plt.show()
```

Marital_Status sales totel ammount

```python
mdf=df.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
ax=sns.barplot(data=mdf,x="Marital_Status",y="Amount",hue="Gender")

plt.show()
```

```python
plt.figure(figsize=(12,6))
ax=sns.countplot(data=df,y="Product_Category")
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```
